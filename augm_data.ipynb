## resize les data augmentées 


```python
import os
import glob
import numpy
from numpy import asarray
from PIL import Image

dst_dir = '/home/chaumeron/stage/essai_cycle_GAN/Image/resize_augm'
os.makedirs(dst_dir, exist_ok=True)

files = glob.glob('/home/chaumeron/stage/essai_cycle_GAN/Image/data_augm4/*.jpg')

shape = 256 

new_im=Image.new('RGB', (256, 256)) # mettre un chiffre un peu plus grand que shape à cause des arrondis

#new_im.save(os.path.join(dst_dir, 'reference'+'.png'))

for f in files:
    img = Image.open(f)
    coeff=(img.width/shape)
    img_resize = img.resize((int(img.width/coeff), int(img.height/coeff)))


    root, ext = os.path.splitext(f)
    basename = os.path.basename(root)

    
   # img_resize.save(os.path.join(dst_dir, basename + '_half' + ext))

    dec_gauche=int((new_im.width-img_resize.width)/2)
    dec_haut=int((new_im.height-img_resize.height)/2)
    new=new_im.copy()
    new.paste(img_resize, (dec_gauche, dec_haut))
    
    #rogne l'image
    gauche=int((new.width-shape)/2)
    haut=int((new.height-shape)/2)

    #print(gauche, haut)

    final=new.crop((gauche, haut, shape+gauche, shape+haut))


    

    root, ext = os.path.splitext(f)
    basename = os.path.basename(root)

    
    final.save(os.path.join(dst_dir, basename + '_half' + ext))
    
    

files
```
