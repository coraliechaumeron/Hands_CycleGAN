```python
import os
import glob
import numpy
from numpy import asarray
from PIL import Image

dst_dir = '/home/chaumeron/Bureau/essai_cycle_GAN/Image/resize'
os.makedirs(dst_dir, exist_ok=True)

files = glob.glob('/home/chaumeron/Bureau/essai_cycle_GAN/Image/*.png')

longueur = []
largeur= []

shape = 3000 

new_im=Image.new('RGB', (3000, 3000)) # mettre un chiffre un peu plus grand que shape à cause des arrondis

new_im.save(os.path.join(dst_dir, 'reference'+'.png'))

for f in files:
    img = Image.open(f)

    longueur.append(img.width)
    largeur.append(img.height)
    coeff=(max(img.width, img.height)/shape)

    print(type(coeff))

    img_resize = img.resize((int(img.width/coeff), int(img.height/coeff)))
    

    dec_gauche=int((new_im.width-img_resize.width)/2)
    dec_haut=int((new_im.height-img_resize.height)/2)
    new=new_im.copy()
    new.paste(img_resize, (dec_gauche, dec_haut))
    
    #rogne l'image
    gauche=int((new.width-shape)/2)
    haut=int((new.height-shape)/2)

    print(gauche, haut)

    final=new.crop((gauche, haut, shape+gauche, shape+haut))

    print(type(final))
    

    root, ext = os.path.splitext(f)
    basename = os.path.basename(root)

    
    final.save(os.path.join(dst_dir, basename + '_half' + ext))
    

files

mL=max(longueur)
ml=max(largeur)

print(mL, ml)


```
